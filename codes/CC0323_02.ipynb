# Golden Section

```python
import math

def f(x):
    return x**4 - 14*x**3 + 60*x**2 - 70*x

def golden_section_search(a, b, tol=1e-3):
    gr = (math.sqrt(5) + 1) / 2  # Golden ratio
    c = b - (b - a) / gr
    d = a + (b - a) / gr
    fc = f(c)
    fd = f(d)

    niters = 0
    while abs(b - a) > tol:
        if fc < fd:
            b = d
            d = c
            fd = fc
            c = b - (b - a) / gr
            fc = f(c)
        else:
            a = c
            c = d
            fc = fd
            d = a + (b - a) / gr
            fd = f(d)
        niters += 1

    x = (a + b) / 2
    print(f'Solution found after {niters} iterations')
    return x, a, b

# Interval [0, 2] and precision tolerance
a = 0
b = 2
tol = 1e-3

# Finding the minimum
xsol, a_final, b_final = golden_section_search(a, b, tol)
xsol, a_final, b_final
```

# Newton-Raphson

```python
# Re-defining the Newton-Raphson method to count iterations
def newton_raphson(f, df, ddf, x0, tol=1e-5, maxit=1000):
    x = x0
    for niters in range(1, maxit + 1):
        ddf_x = ddf(x)
        if ddf_x <= 0:
            raise Exception('The function is not convex')
        df_x = df(x)
        x_next = x - df_x / ddf_x
        print(f'error df(x) {df_x}')
        if abs(x_next - x) < tol:
            return x_next
        x = x_next
    print(f'Solution found after {niters} iterations')
    return x  # Return the last estimate
```

```python
import sympy as sp

# Define the symbolic functions and convert to numerical functions
x = sp.symbols('x')
f_sym = x**4 - 14*x**3 + 60*x**2 - 70*x
df_sym = sp.diff(f_sym, x)
ddf_sym = sp.diff(df_sym, x)

f = sp.lambdify(x, f_sym, 'numpy')
df = sp.lambdify(x, df_sym, 'numpy')
ddf = sp.lambdify(x, ddf_sym, 'numpy')

# Initial guess at the midpoint of the interval [0, 2]
x0 = 1

# Finding the minimum with iteration count
xsol = newton_raphson(f, df, ddf, x0, tol=1e-8)

xsol
```
