```python
import matplotlib.pyplot as plt
import numpy as np

# Função 2: f(x) = (x-1)^2 - 2
x = np.linspace(-1, 2, 400)
y = (x-1)**2 - 2

# Plotando a Função 2
fig = plt.figure(figsize=(8, 4))
plt.plot(x, y, label=r'$f(x) = (x-1)^2 - 2$')
plt.scatter([0], [-1], color='red', label=r'Minimizador $x^* = 0, f(x^*) = -1$')
# line with large width
plt.plot([-1, 0], [-2, -2], color='green', linewidth=3, label=r'$\Omega$')
plt.scatter([-1, 0], [-2, -2], color='green', linewidth=3)
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.legend()
plt.grid(True)
plt.ylim(-2.1, 2)
plt.show()

# save figure
fig.savefig('../slides/CC0323_02/example1.png', dpi=300, bbox_inches='tight');
```

```python
# Função 1: f(x) = x_1^2 + (x_2-1)^2
x1 = np.linspace(-2, 2, 400)
x2 = np.linspace(-1, 3, 400)
X1, X2 = np.meshgrid(x1, x2)
F = X1**2 + (X2-1)**2

# Plotando a Função 1
fig = plt.figure(figsize=(6, 6))
contours = plt.contour(X1, X2, F, levels=np.logspace(-2, 3, 20), cmap='viridis')
plt.clabel(contours, inline=True, fontsize=8)
# plt.scatter([0], [1], color='red', label=r'Minimizador $x^* = (0, 1), f(x^*) = 0$')
plt.title('Curvas de Nível')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.legend()
plt.grid(True)
plt.show()

# save figure
fig.savefig('../slides/CC0323_02/level_curve.png', dpi=300, bbox_inches='tight');

from mpl_toolkits.mplot3d import Axes3D

# Criando o gráfico 3D para a Função 1
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X1, X2, F, cmap='viridis', edgecolor='none', rstride=1, cstride=1, alpha=0.8)
ax.scatter([0], [1], [0], color='red', s=50, label=r'Minimizador $x^* = (0, 1), f(x^*) = 0$')
# ax.set_title('Exemplo 1')
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('$f(x)$')
ax.legend()
fig.colorbar(surf, shrink=0.5, aspect=5)  # Adiciona a barra de cores para referência de valores de f(x)
plt.show()

# save figure
fig.savefig('../slides/CC0323_02/example2_3d.png', dpi=300, bbox_inches='tight');
```
